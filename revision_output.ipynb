# HR gen puro

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test.csv', delimiter=';')

df
```

```python
df.columns
```

```python
mean_cols = [col for col in df.columns if 'mean' in col]
max_cols = [col for col in df.columns if 'max' in col]
minutes = range(1, 13)

plt.figure(figsize=(10, 6))
plt.plot(minutes, df[mean_cols].iloc[0], marker='o', label='HR Mean')
plt.plot(minutes, df[max_cols].iloc[0], marker='x', label='HR Max')
plt.xlabel('Minute')
plt.ylabel('Heart Rate')
plt.title('Heart Rate Mean and Max per Minute')
plt.legend()
plt.grid(True)
plt.show()
```
