```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data
df = pd.read_csv('test.csv', sep=';')

# Convert Segment labels to numeric indices
df['Segment'] = df['Segment'].str.replace('Seg_', '').astype(int)

# Plot Mean and Max statistics per segment for each Event|Channel
for stat in ['Mean', 'Max']:
    fig, ax = plt.subplots()
    for (event, channel), group in df.groupby(['Event', 'Channel']):
        ax.plot(group['Segment'], group[stat], label=f"{event} | {channel}")
    ax.set_xlabel('Segment')
    ax.set_ylabel(stat)
    ax.set_title(f"{stat} by Segment")
    ax.legend()
    plt.tight_layout()
    plt.show()
```

```python
# HR gen puro
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df2 = pd.read_csv('test2.csv', delimiter=';')

mean_cols2 = [col for col in df2.columns if 'mean' in col]
print(len(mean_cols2))
max_cols2 = [col for col in df2.columns if 'max' in col]
print(len(max_cols2))
minutes2 = range(0,df2.shape[1]//2)
print(minutes2)

plt.figure(figsize=(10, 6))
plt.plot(minutes2, df2[mean_cols2].iloc[0], marker='o', label='HR Mean')
plt.plot(minutes2, df2[max_cols2].iloc[0], marker='x', label='HR Max')
plt.xlabel('Minute')
plt.ylabel('Heart Rate')
plt.title('Heart Rate Mean and Max per Minute')
plt.legend()
plt.grid(True)
plt.show()
```
